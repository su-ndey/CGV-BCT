```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

1. Implement fixed-point scaling of a line about its starting point.

```python

def scale_line_fixed_point(x1, y1, x2, y2, sx, sy):
    x2_new = x1 + sx * (x2 - x1)
    y2_new = y1 + sy * (y2 - y1)
    return x2_new, y2_new

x1, y1 = 2, 3
x2, y2 = 8, 7
sx, sy = 1.5, 2
x2_new, y2_new = scale_line_fixed_point(x1, y1, x2, y2, sx, sy)

print(f"Original line: Start=({x1},{y1}), End=({x2},{y2})")
print(f"Scaled line: Start=({x1},{y1}), End=({x2_new:.2f},{y2_new:.2f})")
plt.figure(figsize=(6,6))
plt.plot([x1, x2], [y1, y2], 'b-o', label='Original Line')
plt.plot([x1, x2_new], [y1, y2_new], 'r-o', label='Scaled Line')
plt.title(f"Fixed-Point Scaling about ({x1},{y1})\nSx={sx}, Sy={sy}")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

2. Change the fixed point to the midpoint of the line and observe the difference

```python

def scale_line_midpoint(x1, y1, x2, y2, sx, sy):
    
    xm = (x1 + x2) / 2
    ym = (y1 + y2) / 2

    x1_new = xm + sx * (x1 - xm)
    y1_new = ym + sy * (y1 - ym)
    x2_new = xm + sx * (x2 - xm)
    y2_new = ym + sy * (y2 - ym)

    return x1_new, y1_new, x2_new, y2_new
x1, y1 = 2, 3
x2, y2 = 8, 7

sx, sy = 1.5, 2

x1_new, y1_new, x2_new, y2_new = scale_line_midpoint(x1, y1, x2, y2, sx, sy)

print(f"Original line: Start=({x1},{y1}), End=({x2},{y2})")
print(f"Scaled line about midpoint: Start=({x1_new:.2f},{y1_new:.2f}), End=({x2_new:.2f},{y2_new:.2f})")

plt.figure(figsize=(6,6))
plt.plot([x1, x2], [y1, y2], 'b-o', label='Original Line')
plt.plot([x1_new, x2_new], [y1_new, y2_new], 'r-o', label='Scaled about Midpoint')
plt.scatter((x1+x2)/2, (y1+y2)/2, color='green', s=80, label='Midpoint')
plt.title(f"Line Scaling about Midpoint\nSx={sx}, Sy={sy}")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

3. Implement pure rotation about the origin using the given rotation matrix.

```python

def rotate_line_origin(x1, y1, x2, y2, angle_deg):
    angle_rad = math.radians(angle_deg)

    x1_new = x1 * math.cos(angle_rad) - y1 * math.sin(angle_rad)
    y1_new = x1 * math.sin(angle_rad) + y1 * math.cos(angle_rad)

    x2_new = x2 * math.cos(angle_rad) - y2 * math.sin(angle_rad)
    y2_new = x2 * math.sin(angle_rad) + y2 * math.cos(angle_rad)

    return x1_new, y1_new, x2_new, y2_new

x1, y1 = 3, 2
x2, y2 = 8, 5

angle = 45

x1_r, y1_r, x2_r, y2_r = rotate_line_origin(x1, y1, x2, y2, angle)

print(f"Original line: Start=({x1},{y1}), End=({x2},{y2})")
print(f"Rotated line by {angle}: Start=({x1_r:.2f},{y1_r:.2f}), End=({x2_r:.2f},{y2_r:.2f})")

plt.figure(figsize=(6,6))
plt.plot([x1, x2], [y1, y2], 'b-o', label='Original Line')
plt.plot([x1_r, x2_r], [y1_r, y2_r], 'r-o', label=f'Rotated Line {angle}')
plt.title("Pure Rotation about Origin")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```
