```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

1. Implement the midpoint circle algorithm as above.

```python

def midpoint_circle(xc, yc, r):
    x_coords = []
    y_coords = []

    x = 0
    y = r
    p = 1 - r

    while x <= y:
        points = [
            ( xc + x, yc + y),
            ( xc - x, yc + y),
            ( xc + x, yc - y),
            ( xc - x, yc - y),
            ( xc + y, yc + x),
            ( xc - y, yc + x),
            ( xc + y, yc - x),
            ( xc - y, yc - x)
        ]
        for px, py in points:
            x_coords.append(px)
            y_coords.append(py)

        x += 1
        if p < 0:
            p += 2*x + 1
        else:
            y -= 1
            p += 2*(x - y) + 1

    return x_coords, y_coords

xc, yc = 0, 0  
r = 10         

x_list, y_list = midpoint_circle(xc, yc, r)

print("Circle coordinates:")
for i in range(len(x_list)):
    print(f"({x_list[i]}, {y_list[i]})")

plt.figure(figsize=(6,6))
plt.plot(x_list, y_list, 'o', color='blue')
plt.title(f"Midpoint Circle: Center=({xc},{yc}), Radius={r}")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.axis('equal')  
plt.show()
```

2. Draw circles with different radii and centres.

```python

def midpoint_circle(xc, yc, r):
    x_coords = []
    y_coords = []

    x = 0
    y = r
    p = 1 - r

    while x <= y:
        points = [
            ( xc + x, yc + y),
            ( xc - x, yc + y),
            ( xc + x, yc - y),
            ( xc - x, yc - y),
            ( xc + y, yc + x),
            ( xc - y, yc + x),
            ( xc + y, yc - x),
            ( xc - y, yc - x)
        ]
        for px, py in points:
            x_coords.append(px)
            y_coords.append(py)

        x += 1
        if p < 0:
            p += 2*x + 1
        else:
            y -= 1
            p += 2*(x - y) + 1

    return x_coords, y_coords

circles = [
    (0, 0, 10),
    (15, 5, 7),
    (-10, -10, 5),
    (5, -12, 8)
]

plt.figure(figsize=(8, 8))
plt.title("Multiple Circles using Midpoint Circle Algorithm")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.axis('equal')  

for i, (xc, yc, r) in enumerate(circles):
    x_list, y_list = midpoint_circle(xc, yc, r)
    print(f"Circle {i+1}: Center=({xc},{yc}), Radius={r}")
    plt.plot(x_list, y_list, 'o', label=f"Circle {i+1}")

plt.legend()
plt.show()
```

3. Use the algorithm to draw concentric circles (target pattern)

```python

def midpoint_circle(xc, yc, r):
    x_coords = []
    y_coords = []

    x = 0
    y = r
    p = 1 - r

    while x <= y:
        points = [
            (xc + x, yc + y),
            (xc - x, yc + y),
            (xc + x, yc - y),
            (xc - x, yc - y),
            (xc + y, yc + x),
            (xc - y, yc + x),
            (xc + y, yc - x),
            (xc - y, yc - x)
        ]
        for px, py in points:
            x_coords.append(px)
            y_coords.append(py)

        x += 1
        if p < 0:
            p += 2*x + 1
        else:
            y -= 1
            p += 2*(x - y) + 1

    return x_coords, y_coords

center_x, center_y = 0, 0
num_circles = 5
radius_step = 5  

plt.figure(figsize=(6,6))
plt.title("Target Pattern: Concentric Circles")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.axis('equal')  

for i in range(1, num_circles + 1):
    r = i * radius_step
    x_list, y_list = midpoint_circle(center_x, center_y, r)
    plt.plot(x_list, y_list, 'o', label=f"r={r}")

plt.legend()
plt.show()
```
