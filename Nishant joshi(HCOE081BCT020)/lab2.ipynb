1. Implement the DDA algorithm as a function that returns x and y coordinate
lists

```python

import matplotlib.pyplot as plt

def dda(x1, y1, x2, y2):
    x_coords = []
    y_coords = []

    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))

    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    for _ in range(steps + 1):
        x_coords.append(round(x))
        y_coords.append(round(y))
        x += x_inc
        y += y_inc

    return x_coords, y_coords

x1, y1 = 2, 3
x2, y2 = 10, 8

x_list, y_list = dda(x1, y1, x2, y2)

print("X coordinates:", x_list)
print("Y coordinates:", y_list)

plt.plot(x_list, y_list, marker='o')
plt.title(f"DDA Line from ({x1},{y1}) to ({x2},{y2})")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

```

```python

def dda(x1, y1, x2, y2):
    x_coords = []
    y_coords = []

    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))

    if steps == 0:  
        return [x1], [y1]

    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    for _ in range(int(steps) + 1):
        x_coords.append(round(x))
        y_coords.append(round(y))
        x += x_inc
        y += y_inc

    return x_coords, y_coords

print("Enter endpoints for the line:")

try:
    x1 = int(input("Enter x1: "))
    y1 = int(input("Enter y1: "))
    x2 = int(input("Enter x2: "))
    y2 = int(input("Enter y2: "))
except ValueError:
    print("Please enter valid integer coordinates.")
    exit()

x_list, y_list = dda(x1, y1, x2, y2)

print("\nGenerated line coordinates:")
print("X coordinates:", x_list)
print("Y coordinates:", y_list)

plt.figure(figsize=(6, 6))
plt.plot(x_list, y_list, marker='o', color='blue')
plt.title(f"DDA Line from ({x1},{y1}) to ({x2},{y2})")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

1. Extend the DDA program to draw a rectangle given two opposite corners.

```python

def dda(x1, y1, x2, y2):
    x_coords = []
    y_coords = []

    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))

    if steps == 0:  
        return [x1], [y1]

    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    for _ in range(int(steps) + 1):
        x_coords.append(round(x))
        y_coords.append(round(y))
        x += x_inc
        y += y_inc

    return x_coords, y_coords


print("Enter two opposite corners of the rectangle:")

try:
    x1 = int(input("Enter x1: "))
    y1 = int(input("Enter y1: "))
    x2 = int(input("Enter x2: "))
    y2 = int(input("Enter y2: "))
except ValueError:
    print("Please enter valid integer coordinates.")
    exit()

x3, y3 = x1, y2  
x4, y4 = x2, y1  

lines = [
    (x1, y1, x3, y3),  
    (x3, y3, x2, y2),  
    (x2, y2, x4, y4),  
    (x4, y4, x1, y1)   
]

plt.figure(figsize=(6, 6))
plt.title(f"Rectangle from ({x1},{y1}) to ({x2},{y2})")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

for i, (xa, ya, xb, yb) in enumerate(lines):
    x_list, y_list = dda(xa, ya, xb, yb)
    print(f"Line {i+1} coordinates:")
    print("  X:", x_list)
    print("  Y:", y_list)
    plt.plot(x_list, y_list, marker='o', color='blue')

plt.show()
```

2. Use DDA to draw the axes of a simple coordinate system (X and Y axes).

```python

def dda(x1, y1, x2, y2):
    x_coords = []
    y_coords = []

    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))

    if steps == 0:  
        return [x1], [y1]

    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    for _ in range(int(steps) + 1):
        x_coords.append(round(x))
        y_coords.append(round(y))
        x += x_inc
        y += y_inc

    return x_coords, y_coords

print("Enter the range for the coordinate system:")

try:
    xmin = int(input("Enter minimum x: "))
    xmax = int(input("Enter maximum x: "))
    ymin = int(input("Enter minimum y: "))
    ymax = int(input("Enter maximum y: "))
except ValueError:
    print("Please enter valid integer coordinates.")
    exit()

x_axis_x, x_axis_y = dda(xmin, 0, xmax, 0)

y_axis_x, y_axis_y = dda(0, ymin, 0, ymax)

print("\nX-axis coordinates:")
print("X:", x_axis_x)
print("Y:", x_axis_y)

print("\nY-axis coordinates:")
print("X:", y_axis_x)
print("Y:", y_axis_y)

plt.figure(figsize=(8, 8))
plt.plot(x_axis_x, x_axis_y, marker='o', color='red', label='X-axis')
plt.plot(y_axis_x, y_axis_y, marker='o', color='green', label='Y-axis')
plt.title("Coordinate System using DDA")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.show()
```
