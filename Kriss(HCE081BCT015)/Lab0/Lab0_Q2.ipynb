```python

# LAB-0
# Q2: Use DDA to draw X-axis and Y-axis

import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))

    if steps == 0:
        return [x1], [y1]

    x_inc = dx / steps
    y_inc = dy / steps
    x, y = x1, y1
    xs, ys = [], []

    for _ in range(steps + 1):
        xs.append(round(x))
        ys.append(round(y))
        x += x_inc
        y += y_inc

    return xs, ys


limit = int(input("Enter axis limit: "))

X, Y = [], []

# X-axis
xs, ys = dda_line(-limit, 0, limit, 0)
X += xs
Y += ys

# Y-axis
xs, ys = dda_line(0, -limit, 0, limit)
X += xs
Y += ys

plt.plot(X, Y, marker='o', color='green')
plt.title("DDA Coordinate Axes")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```
