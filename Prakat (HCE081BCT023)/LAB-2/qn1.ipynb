```python
# LAB-2
# Q1: Implement Bresenham's Line Drawing Algorithm

import matplotlib.pyplot as plt

def bresenham_line(x1, y1, x2, y2):
    x_points = []
    y_points = []

    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    labelx = 1 if x1 < x2 else -1
    labely = 1 if y1 < y2 else -1

    x, y = x1, y1

    if dx >= dy:
        pk = 2 * dy - dx

        for _ in range(dx + 1):
            x_points.append(x)
            y_points.append(y)

            if pk >= 0:
                y += labely
                pk += 2 * (dy - dx)
            else:
                pk += 2 * dy

            x += labelx

    else:
        pk = 2 * dx - dy

        for _ in range(dy + 1):
            x_points.append(x)
            y_points.append(y)

            if pk >= 0:
                x += labelx
                pk += 2 * (dx - dy)
            else:
                pk += 2 * dx

            y += labely

    return x_points, y_points


x1 = int(input("Enter the value of x1: "))
y1 = int(input("Enter the value of y1: "))
x2 = int(input("Enter the value of x2: "))
y2 = int(input("Enter the value of y2: "))

x_pts, y_pts = bresenham_line(x1, y1, x2, y2)

plt.plot(x_pts, y_pts, marker='o', color='green')
plt.title("Bresenham's Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
