```python
import matplotlib.pyplot as plt

# DDA algorithm function
def dda(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0
    steps = max(abs(dx), abs(dy))
    x_inc = dx / steps
    y_inc = dy / steps
    x = x0
    y = y0
    x_coords = []
    y_coords = []
    for _ in range(int(steps) + 1):
        x_coords.append(round(x))
        y_coords.append(round(y))
        x += x_inc
        y += y_inc
    return x_coords, y_coords

# 1 Draw sample lines with different slopes
sample_lines = [
    (0, 0, 10, 5),   # m < 1
    (0, 0, 5, 10),   # m > 1
    (0, 0, 10, 0),   # horizontal
    (0, 0, 0, 10),   # vertical
    (0, 0, 10, -5)   # negative slope
]

plt.figure(figsize=(6,6))
for line in sample_lines:
    x0, y0, x1, y1 = line
    x_coords, y_coords = dda(x0, y0, x1, y1)
    plt.plot(x_coords, y_coords, marker='o', label=f"{(x0,y0)}{(x1,y1)}")

plt.title("DDA Line Drawing - Sample Slopes")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.show()  

# 2 Take user input for custom line
print("\nNow, enter coordinates for your custom line:")
try:
    x0 = int(input("Enter x0: "))
    y0 = int(input("Enter y0: "))
    x1 = int(input("Enter x1: "))
    y1 = int(input("Enter y1: "))
except ValueError:
    print("Invalid input! Please enter integers only.")
    exit()

x_coords, y_coords = dda(x0, y0, x1, y1)
plt.figure(figsize=(6,6))
plt.plot(x_coords, y_coords, marker='o', color='red')
plt.title(f"DDA Line from ({x0},{y0}) to ({x1},{y1})")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt


def dda(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0
    steps = max(abs(dx), abs(dy))
    x_inc = dx / steps
    y_inc = dy / steps
    x = x0
    y = y0
    x_coords = []
    y_coords = []
    for _ in range(int(steps) + 1):
        x_coords.append(round(x))
        y_coords.append(round(y))
        x += x_inc
        y += y_inc
    return x_coords, y_coords

# -----------------------
# Task 1: Draw rectangle given two opposite corners
# -----------------------
print("Enter two opposite corners of the rectangle:")
x0 = int(input("x0: "))
y0 = int(input("y0: "))
x1 = int(input("x1: "))
y1 = int(input("y1: "))

# Rectangle corners
x2, y2 = x0, y1
x3, y3 = x1, y0

# Draw rectangle edges
plt.figure(figsize=(6,6))
edges = [
    (x0, y0, x2, y2),
    (x2, y2, x1, y1),
    (x1, y1, x3, y3),
    (x3, y3, x0, y0)
]

for edge in edges:
    ex, ey = dda(*edge)
    plt.plot(ex, ey, marker='o')

plt.title("Rectangle using DDA")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

# -----------------------
# Task 2: Draw coordinate axes
# -----------------------
plt.figure(figsize=(6,6))

# X axis from -50 to 50
x_axis, y_axis = dda(-50, 0, 50, 0)
plt.plot(x_axis, y_axis, color='blue', label='X-axis')

# Y axis from -50 to 50
x_axis, y_axis = dda(0, -50, 0, 50)
plt.plot(x_axis, y_axis, color='green', label='Y-axis')

plt.title("Simple Coordinate System using DDA")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.show()
```
