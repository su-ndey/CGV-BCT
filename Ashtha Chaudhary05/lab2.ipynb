```python
# Question 1: DDA Line Algorithm 
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    xes, yes = [], []
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x = x1
    y = y1
    for _ in range(steps + 1):
        xes.append(round(x))
        yes.append(round(y))
        x += x_inc
        y += y_inc
    return xes, yes

def plot_dda(x1, y1, x2, y2, title="DDA Line"):
    xes, yes = dda_line(x1, y1, x2, y2)
    plt.plot(xes, yes, marker='o')
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.axis("equal")

# Run Question 1
plt.figure(figsize=(6,6))
plot_dda(2, 3, 15, 9, "Question 1: DDA Line")
plt.show()
```

```python
# Question 4: Rectangle Using Two Opposite Corners 
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    xes, yes = [], []
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x = x1
    y = y1
    for _ in range(steps + 1):
        xes.append(round(x))
        yes.append(round(y))
        x += x_inc
        y += y_inc
    return xes, yes

def rectangle_dda(x1, y1, x3, y3):
    x2, y2 = x3, y1
    x4, y4 = x1, y3

    plt.figure(figsize=(6,6))
    for (xa, ya, xb, yb) in [(x1, y1, x2, y2),
                             (x2, y2, x3, y3),
                             (x3, y3, x4, y4),
                             (x4, y4, x1, y1)]:
        xes, yes = dda_line(xa, ya, xb, yb)
        plt.plot(xes, yes, marker='o')

    plt.title("Question 4: Rectangle using DDA")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.axis("equal")
    plt.show()

# Run Question 4
rectangle_dda(50, 50, 65, 65)
```
