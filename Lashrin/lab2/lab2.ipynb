```python
# Extend the DDA program to draw a rectangle given two opposite corners. 
import matplotlib.pyplot as plt
import math

def dda_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))

    x_inc = dx / steps
    y_inc = dy / steps

    x, y = x1, y1
    x_points = []
    y_points = []

    for _ in range(steps + 1):
        x_points.append(round(x))
        y_points.append(round(y))
        x += x_inc
        y += y_inc

    plt.plot(x_points, y_points, 'k')

x1, y1 = map(int, input("Enter first corner (x1 y1): ").split())
x2, y2 = map(int, input("Enter opposite corner (x2 y2): ").split())

x3, y3 = x2, y1
x4, y4 = x1, y2

# Draw rectangle
dda_line(x1, y1, x3, y3)
dda_line(x3, y3, x2, y2)
dda_line(x2, y2, x4, y4)
dda_line(x4, y4, x1, y1)

plt.title("Rectangle using DDA Algorithm")
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```

```python
#Use DDA to draw the axes of a simple coordinate system (X and Y axes).
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))
    x_inc = dx / steps
    y_inc = dy / steps

    x, y = x1, y1
    x_points, y_points = [], []

    for _ in range(steps + 1):
        x_points.append(round(x))
        y_points.append(round(y))
        x += x_inc
        y += y_inc

    plt.plot(x_points, y_points, 'k')

L = 50

dda_line(-L, 0, L, 0)

dda_line(0, -L, 0, L)

plt.axhline(0)
plt.axvline(0)
plt.title("Coordinate Axes using DDA Algorithm")
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```
