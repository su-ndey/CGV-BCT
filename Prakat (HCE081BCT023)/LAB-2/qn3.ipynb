```python
# LAB-2
# Q3: Compare number of additions and multiplications used by DDA and Bresenham

def compare_operations(x1, y1, x2, y2):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    steps = max(dx, dy)

    # DDA operations
    dda_additions = steps * 2
    dda_multiplications = steps * 2

    # Bresenham operations
    bres_additions = steps * 3
    bres_multiplications = 0

    print("Line from ({}, {}) to ({}, {})".format(x1, y1, x2, y2))
    print("\nDDA Algorithm:")
    print("Additions:", dda_additions)
    print("Multiplications:", dda_multiplications)

    print("\nBresenham Algorithm:")
    print("Additions:", bres_additions)
    print("Multiplications:", bres_multiplications)

compare_operations(2, 3, 15, 9)
```
