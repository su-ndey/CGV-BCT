```python
import matplotlib.pyplot as plt

def midpoint_circle(xc, yc, r):
    x, y = 0, r
    d = 1 - r
    x_coords, y_coords = [], []
    while x <= y:
        # 8-way symmetry
        pts = [(xc+x, yc+y), (xc-x, yc+y), (xc+x, yc-y), (xc-x, yc-y),
               (xc+y, yc+x), (xc-y, yc+x), (xc+y, yc-x), (xc-y, yc-x)]
        for px, py in pts:
            x_coords.append(px)
            y_coords.append(py)
        if d < 0:
            d += 2*x + 3
        else:
            d += 2*(x - y) + 5
            y -= 1
        x += 1
    return x_coords, y_coords

# Draw circles with different radii and centers
circles = [(0,0,10), (15,15,7), (-10,-5,5)]
plt.figure(figsize=(6,6))
for c in circles:
    xc,yc,r = c
    x_pts, y_pts = midpoint_circle(xc,yc,r)
    plt.scatter(x_pts, y_pts, s=10)

# Concentric circles (target pattern)
xc, yc = 20, 0
for r in [3,6,9,12]:
    x_pts, y_pts = midpoint_circle(xc,yc,r)
    plt.scatter(x_pts, y_pts, s=10)

plt.title("Midpoint Circle Algorithm")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.axis('equal')
plt.show()
```
