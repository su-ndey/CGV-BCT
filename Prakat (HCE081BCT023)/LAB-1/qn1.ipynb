```python
# LAB-1
# Q1: Extend DDA to draw a rectangle given two opposite corners

import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))

    if steps == 0:
        return [x1], [y1]

    x_inc = dx / steps
    y_inc = dy / steps
    x, y = x1, y1
    xs, ys = [], []

    for _ in range(steps + 1):
        xs.append(round(x))
        ys.append(round(y))
        x += x_inc
        y += y_inc

    return xs, ys



x3, y3 = 10, 20
x4, y4 = 30, 40
X, Y = [], []

for a, b, c, d in [
    (x1, y1, x3, y3),
    (x3, y3, x2, y2),
    (x2, y2, x4, y4),
    (x4, y4, x1, y1)
]:
    xs, ys = dda_line(a, b, c, d)
    X += xs
    Y += ys

plt.plot(X, Y, marker='o', color='blue')
plt.title("DDA Rectangle Drawing")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```
