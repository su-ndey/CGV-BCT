```python
import numpy as np
import matplotlib.pyplot as plt

# x values from 0 to 2Ï€
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# plot the sine wave
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()  # this is important to display the graph
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Example usage
num = int(input("Enter a number: "))
print(f"The factorial of {num} is {factorial(num)}")
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Print primes from 1 to 100
for num in range(1, 101):
    if is_prime(num):
        print(num, end=" ")
```
