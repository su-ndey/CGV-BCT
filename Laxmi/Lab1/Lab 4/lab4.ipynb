```python
#1. Implement the midpoint circle algorithm
import matplotlib.pyplot as plt

def plot_circle_points(xc,yc,x,y,xes,yes):
    pts=[
        ( x+xc, y+yc),
        (-x+xc, y+yc),
        ( x+xc,-y+yc),
        (-x+xc,-y+yc),
        ( y+xc, x+yc),
        (-y+xc, x+yc),
        ( y+xc,-x+yc),
        (-y+xc,-x+yc)
    ]
    for px,py in pts:
        xes.append(px)
        yes.append(py)

def midpoint_circle(r,xc=0,yc=0):
    x=0
    y=r
    p=1-r
    xes,yes=[],[]
    plot_circle_points(xc,yc,x,y,xes,yes)
    while x<y:
        x+=1
        if p<0:
            p=p+2*x+1
        else:
            y-=1
            p=p+2*(x-y)+1
        plot_circle_points(xc,yc,x,y,xes,yes)
    return xes,yes

def plot_midpoint_circle(r,xc=0,yc=0):
    xes,yes=midpoint_circle(r,xc,yc)
    plt.figure(figsize=(6,6))
    plt.scatter(xes,yes,marker='.',color='red')
    plt.title("Midpoint Circle Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.axis('equal')
    plt.show()

plot_midpoint_circle(20,0,20)
```

```python
#3. Use the algorithm to draw concentric circles (target pattern).
import matplotlib.pyplot as plt

radii = [15, 20, 30, 40, 50]
colors = ['blue', 'orange', 'black', 'red', 'green']

fig, ax = plt.subplots()

for r, c in zip(radii, colors):
    circle = plt.Circle((0, 0), r, color=c, fill=False, linewidth=2)
    ax.add_patch(circle)

ax.set_aspect('equal', adjustable='box')
ax.set_xlim(-100, 70)
ax.set_ylim(-100, 70)

plt.grid(True)
plt.title("Concentric Circles")
plt.show()
```
